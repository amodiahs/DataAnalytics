We will perform EDA (Exploratory Data Analysis) on the weater data that we extracted from Estes Park Weather side through Web Scraping, transformed it to Pandas Data Frame and loaded it to sqlite database

We will use below Python modules
* Pandas for data exploration
* Matplotlib and Seaborn for data visualization

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sbn

import sqlite3

%matplotlib inline
```

Lets read the data that we loaded to sqlite in previous exercise. If anyone missed that then you can download the ETL Jupyter Notebook and sqlite database file from below path of Git Repository:
https://github.com/amodiahs/DataAnalytics/tree/main/EstesParkWeather

```python
connection = sqlite3.connect("weather_data.db")
weather_df = pd.read_sql_query("SELECT * from Weather", connection)
connection.close()
```

We will now use pandas describe function which is a handy utility to know important basic statistical values of each component of the data

```python
weather_df.describe()
```

```python
max_avg_temp = weather_df["average_temperature"].max()
print(f'Max avg temperature recorded is {max_avg_temp} Faranheit')
```

```python
min_avg_temp = weather_df["average_temperature"].min()
print(f'Min avg temperature recorded is {min_avg_temp} Faranheit')
```

```python
mean_avg_temp = weather_df["average_temperature"].mean()
print(f'Mean avg temperature recorded is {mean_avg_temp} Faranheit')
```

```python
median_avg_temp = weather_df["average_temperature"].median()
print(f'Median avg temperature recorded is {median_avg_temp} Faranheit')
```

In the next few cells we would do some graphical analysis of the data

```python
weather_df['average_temperature'].hist(bins=5)
```

```python
sbn.boxplot(y=weather_df['average_temperature'], orient='v')
```

```python
sbn.violinplot(y=weather_df['average_temperature'], orient='v')
```

```python
plt.figure(figsize=(15, 6))
plt.xticks(rotation=65)
plt.plot(weather_df['date'], weather_df['average_temperature'])
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=65)
plt.plot(weather_df['date'], weather_df['average_temperature'])
plt.plot(weather_df['date'], weather_df['maximum_temperature'])
plt.plot(weather_df['date'], weather_df['minimum_temperature'])
plt.legend(['Average Temperature', 'Maximum Temperature', 'Minumum Temperature'])
```
